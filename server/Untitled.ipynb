```python
def beautify(word):
    return word.replace("_", " ").strip()
```

```python
SYMPTOMS = list(mlb.classes_)
beautified = [beautify(word) for word in SYMPTOMS]
SYMPTOMS = beautified

# Create matcher
matcher = PhraseMatcher(nlp.vocab, attr="LOWER")
patterns = [nlp.make_doc(symptom) for symptom in SYMPTOMS]
matcher.add("SYMPTOM", patterns)

def extract_symptoms(text):
    doc = nlp(text)
    matches = matcher(doc)
    matched_symptoms = list(set([doc[start:end].text.lower() for match_id, start, end in matches]))
    return matched_symptoms

# Test
text = "mild fever I am have acidity"
print(extract_symptoms(text))
```
